```python
with open("the-verdict.txt","r",encoding="utf-8") as f:
    raw_text=f.read()
print("Total characters: ",len(raw_text))
```

```python
raw_text
```

```python
##Without whitespaces
import re
preprocessed=re.split(r'([,.?_!"()\']|--|\s)',raw_text)
preprocessed=[item.strip() for item in preprocessed if item.strip()]
len(preprocessed)
```

```python
preprocessed[:40]
```

```python
##COnvert this tokens to tokenids
allwords=sorted(list(set(preprocessed)))
print("Total Vocab: ",len(allwords))
vocab={token:index for index,token in enumerate(allwords)}
print(vocab)
```

```python
vocab.items()
```

```python
##Tokenizetr
class simpleTokenizer:
    def __init__(self,vocab):
        self.str_to_int=vocab
        self.int_to_str={i:s for s,i in vocab.items()}

    def encode(self,text):
        preprocessed=preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed=[item.strip() for item in preprocessed if item.strip()]
        ids=[self.str_to_int[s] for s in preprocessed]
        return ids
    def decode(self,ids):
        text=" ".join([self.int_to_str[i] for i in ids] )
        return text
tokenizer=simpleTokenizer(vocab)
text=""""It's the last he painted, you know," 
           Mrs. Gisburn said with pardonable pride."""
ids=tokenizer.encode(text)
print(ids)
text=tokenizer.decode(ids)
print(text)
```

```python
allwords=sorted(list(set(preprocessed)))
allwords    .extend(["<UNK>","<EOD>"])
print("Total Vocab: ",len(allwords))
vocab={token:index for index,token in enumerate(allwords)}
print(vocab)
```

```python
##Tokenizer with <unk>
class simpleTokenizerV2:
    def __init__(self,vocab):
        self.str_to_int=vocab
        self.int_to_str={i:s for s,i in vocab.items()}
    def encode(self,text):
        preprocessed=re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed=[item.strip() for item in preprocessed if item.strip()]
        preprocessed=[item if item in self.str_to_int else "<UNK>" for item in preprocessed]
        ids=[self.str_to_int[s] for s in preprocessed]
        return ids
    def decode(self,ids):
        text=" ".join([self.int_to_str[i] for i in ids] )
        return text
tokenizer=simpleTokenizerV2(vocab)
text="""Hello how are you!"""
ids=tokenizer.encode(text)
print(ids)
text=tokenizer.decode(ids)
print(text)   
        
```
